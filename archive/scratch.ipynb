```python
import json
store = json.load(open('/Users/finlayma84/Downloads/Citation Project/repertoire.json'))
print(f"Loaded {len(store)} records.")
```

```python
my_pentecost_preludes = []
for r in store:
    if r['season'] == 'Pentecost' and r['slot'] == 'Prelude':
        my_pentecost_preludes.append(r)

my_pentecost_preludes

print(f"Found {len(my_pentecost_preludes)} Pentecost preludes.\n")
for r in my_pentecost_preludes:
    print(f"  {r['title']}")
```
